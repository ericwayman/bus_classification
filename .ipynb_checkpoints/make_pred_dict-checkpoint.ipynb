```python
#make "pred_dict", a dict of the form {stop_id: {trip_id: sched_pred_trip_id}}
#the keys are stop_id's, and the values are dicts that for each stop value link the trip_id at that stop to the trip_id
#for the scheduled trip_id of the predecessor
#pred_dict is used to let us find the scheduled predecessor for any bus.

#imports
import pandas as pd
import sqlite3
import pandas.io.sql as sql
import numpy as np
```

```python
#make sql connection
connection = sqlite3.connect('predecessor.db')
sched_df = sql.read_frame(sql='SELECT * FROM sched_arrival ORDER BY service_id,stop_id,sched_arrival_time',con=connection)
```

```python

sched_df.head(20)
```

```python
#add columns for the sched_pred_trip_id and sched_pred_stop_id
#TO DO: convert trip_id back to string.  Pandas is inferring the data type as a float
sched_df['sched_pred_trip_id']= sched_df['trip_id'].shift(1)
#sched_df['sched_pred_stop_id']= sched_df['stop_id'].shift(1)
sched_df.head(10)
```

```python
#check that the predecessor has the same stop_id and service id
num_rows = len(sched_df.index)
for i in np.arange(1,num_rows):
    if (sched_df['stop_id'][i] != sched_df['stop_id'][i-1] or sched_df['service_id'][i] != sched_df['service_id'][i-1]):
        sched_df['sched_pred_trip_id'][i] = float('NaN')
len(sched_df['sched_pred_trip_id'] == float('NaN'))
```

```python
stop_id_list = list(set(sched_df['stop_id']))
print stop_id_list
```

```python
#is there a better way to do this using groupby?
dict_list=[]
for s in stop_id_list:
    temp_df = sched_df.ix[sched_df['stop_id']==s]
    trip_ids = list(temp_df['trip_id'])
    pred_trip_ids = list(temp_df['sched_pred_trip_id'])
    dict_list.append(dict(zip(trip_ids,pred_trip_ids)))
```

```python
d = dict(zip([1,2,3],['a','b','c']))
print d
```

```python

pred_dict = dict(zip(stop_id_list,dict_list))
```

```python
#make df for observed data
obs_df= sql.read_frame(sql='SELECT trip_id,service_id,bus_id,arrival_time,stop_id,estimated_boardings,dwell_time, headway, delay, demand FROM bus_data', con=connection)

#convert arrival time to a date time object
obs_df['arrival_time'] = pd.to_datetime(obs_df['arrival_time'])
```

```python
#sort obs_df by stop then by service id then by arrival_time
obs_df = obs_df.sort(['service_id','stop_id','arrival_time'],ascending= True)
num_rows = len(obs_df.index)
#reset the index
obs_df.index = pd.Series(np.arange(num_rows))
```

```python
#make a shifted copy of the data frame to act as the observed predecessor data
df2 = obs_df.shift(1)
new_cols = ["obs_pred_"+x for x in list(obs_df.columns)]
new_cols
df2.columns = new_cols
```

```python
#join the two data frames
pred_df = obs_df.join(df2,how='outer')
pred_df.head(10)
```

```python
#check if stop_id and obs_pred_stop_id correspond.  Place 'NaN' in for value of obs_pred_service_id for arrivals where there is no true predecessor
for i in np.arange(len(pred_df.index)):
    if(pred_df['stop_id'][i] !=pred_df['obs_pred_stop_id'][i]):
        pred_df['obs_pred_stop_id'][i] = float('NaN')
    elif(pred_df['service_id'][i]!= pred_df['obs_pred_service_id'][i]):
        pred_df['obs_pred_service_id'][i] = float('NaN')
```

```python
#crude way to create 'sched_pred_trip_id' column
#for more efficient way, use groupedby?
pred_df['sched_pred_trip_id'] = pd.Series([])
for i in np.arange(len(pred_df.index)):
    pred_df['sched_pred_trip_id'][i] = pred_dict[pred_df['stop_id'][i]][pred_df['trip_id'][i]]
```

```python
pred_df.head(20)
```

```python
#drop data points that don't have obs_pred_trip_id or sched_pred_trip_id or other points with missing data
pred_df = pred_df.dropna(axis=0)
#reset the index
pred_df.index = pd.Series(np.arange(len(pred_df.index)))
```

```python
#initialize the column of responses as an empty Series
pred_df["response"]=pd.Series([])
pred_df.head(10)
for i in np.arange(len(pred_df.index)):
    if (pred_df['sched_pred_trip_id'][i]==pred_df['obs_pred_trip_id'][i]):
        pred_df['response'][i] = 1
    else:
        pred_df['response'][i] = 0
```

```python
pred_df[['response','arrival_time']].head(20)
```

```python
print('False predecessors ' + str(sum(pred_df['response'] == 0)))
print('True predecessors ' + str(sum(pred_df['response'] == 1)))
pred_df.count()
#print type(pred_df['obs_pred_trip_id'][2])
```

```python
sql.write_frame(frame=pred_df,name = 'prediction_data',con=connection,if_exists='replace')
```
