```python
#imports
import pandas as pd
import sqlite3
import pandas.io.sql as sql
import numpy as np
```

```python
#make sql connection
connection = sqlite3.connect('predecessor.db')
#import the database as a pandas df
df= sql.read_frame(sql='SELECT trip_id,service_id,bus_id,arrival_time,stop_id,estimated_boardings,dwell_time FROM bus_data', con=connection)

#convert arrival time to a date time object
df['arrival_time'] = pd.to_datetime(df['arrival_time'])
```

```python
df.head(5)
```

```python
df = df.sort(['stop_id','service_id','arrival_time'],ascending= True)
num_rows = len(df.index)
df.index = pd.Series(np.arange(num_rows))
```

```python
df.head(20)
```

```python
type(df.columns)
```

```python
list(df.columns)
```

```python
#make a shifted copy of the data frame to act as the observed predecessor data
df2 = df.shift(1)
```

```python
#relabel the column names
new_cols = ["obs_pred_"+x for x in list(df.columns)]
new_cols
df2.columns = new_cols
```

```python
df2.head(5)
```

```python
#join the two data frames
pred_df = df.join(df2,how='outer')
```

```python
pred_df.head(10)
```

```python
#remove rows that don't correspond to the observed predecessor

#check if stop_id and obs_pred_stop_id correspond
for i in np.arange(len(pred_df.index)):
    if(pred_df['stop_id'][i] !=pred_df['obs_pred_stop_id'][i]):
        pred_df['obs_pred_stop_id'][i] = float('NaN')
    elif(pred_df['service_id'][i]!= pred_df['obs_pred_service_id'][i]):
        pred_df['obs_pred_service_id'][i] = float('NaN')
```
